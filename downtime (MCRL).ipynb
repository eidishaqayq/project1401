```python
import numpy as np
import random
def downtimeofpolicy2(tf=(2,6.5,2.5,6,5,3.5,3,3.5),time_interval=5,
                      weibull_scale=(2365.08,996.88,713.55,1406.84,343.76,3933.12,828.19,2040.95),
                     weibull_shape=(414.16,109.25,79.81,115.21,169.81,143.60,43.83,296.48),
                     running_time=100000,
                     optimal_replace=(2355,985,695,1375,340,3840,770,2025),
                     tp=(0.4,5.42,0.625,0.857,1.25,0.7,0.429,0.875)):
    age= np.zeros(8)
    downtime=np.zeros(1000)
    num_failed=np.zeros(1000)
    num_replace=np.zeros(1000)
    random_timefailure = np.zeros(8)
    for j in range(1000):
        for i in range(running_time//time_interval):
            for a in range(8):
                random_timefailure[a] = random.weibullvariate(weibull_scale[a],weibull_shape[a])
               
                if (random_timefailure[a] > age[a] and optimal_replace[a] > age[a]) :
                    age[a] +=time_interval
                
                if (random_timefailure[a] <= age[a] and optimal_replace[a] > age[a]) :
                    age[a] = 0
                    downtime[j] += tf[a]
                    num_failed[j] += 1
            
                if (random_timefailure[a] > age[a] and optimal_replace[a] <= age[a]) : 
                    age[a] = 0
                    downtime[j] += tp[a]
                    num_replace[j] +=1
             
                if (random_timefailure[a] < age[a] and optimal_replace[a] < age[a]) :
                    if (random_timefailure[a] > optimal_replace[a]) :
                        age[a] = 0
                        downtime[j] += tp[a]
                        num_replace[j] += 1
                    else:
                        age[a] = 0
                        downtime[j] += tf[a]
                        num_failed[j] +=1
            
                
    return print("downtime=",round(sum(downtime)/1000,2),
                 "number of failure=",round(sum(num_failed)/1000,0),
                 "number of replaced=",round(sum(num_replace)/1000,0))
                    
    
```
